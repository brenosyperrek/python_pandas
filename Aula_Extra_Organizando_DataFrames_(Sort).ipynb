```python
import pandas as pd
```

```python
data = [[1, 2, 3],[4, 5, 6],[7, 8, 9]]
data
```

```python
df = pd.DataFrame(data, list('321'), list('ZYX'))
df
```

```python
df.sort_index(inplace = True)
```

```python
df
```

```python
df.sort_index(inplace = True, axis = 1)
df
```

```python
df.sort_values(by = 'X', inplace = True)
df
```

```python
df.sort_values(by = '3', inplace = True, axis = 1)
df
```
