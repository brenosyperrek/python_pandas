# Relatório de Análise 6

## Criando Novas Variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados_residencial.csv', sep = ';')
dados.head(10)
```

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU'] # Calculando valor total (Aluguel+Condomínio+IPTU)
dados.head(10)
```

```python
dados['Valor M2'] = (dados['Valor'] / dados['Area']).round(2) # Calculando valor por metro quadrado
dados.head()
```

```python
dados['Valor Bruto M2'] = (dados['Valor Bruto'] / dados['Area']).round(2) # Calculando valor bruto por metro quadrado
dados.head()
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x : 'Casa' if x in casa else 'Apartamento')
dados
```

## Excluir Variáveis

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor M2', 'Valor Bruto', 'Valor Bruto M2']])
dados_aux.head(10)
```

```python
del dados_aux['Valor Bruto']
dados_aux.head()
```

```python
dados_aux.pop('Valor Bruto M2')
dados_aux.head()
```

```python
dados_aux.drop(['Valor M2'], axis = 1, inplace = True)
dados_aux.head()
```

```python
dados.drop(['Valor Bruto', 'Valor Bruto M2'], axis = 1, inplace = True)
dados.head()
```

```python
dados.to_csv('dados_residencial.csv', sep = ';', index = False)
```
