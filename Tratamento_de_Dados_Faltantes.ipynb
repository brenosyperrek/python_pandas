## Tratamento de Dados Faltantes (Continuação)

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
selecao
```

```python
A = dados.shape[0]
dados = dados[~selecao]
B = dados.shape[0]
A - B
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
dados.fillna(0, inplace = True) # dados = dados.fillna({'Condominio': 0, 'IPTU': 0})
dados.info()
```

```python
dados.to_csv('dados_residencial.csv', sep = ';', index = False)
```
